# Assignment 3 - Better Integrators

```python
from numpy import *
from math import pi, cos, sqrt

# Physical constants

G = 9.81 # Acceleration due to gravity

# Program constants

POSITION = 0
VELOCITY = 1

# Integrator(s)

def euler(x, f, dt):
    """
    x: one dimensional numpy array
    f: function to apply
    dt: time step
    """
    return x + f(x) * dt

def euler_richardson(x, f, dt):
    x_mid = x + f(x) * 0.5 * dt
    return x + f(x_mid) * dt

def runge_kutta(x, f, dt):
    k1 = f(x) * dt
    k2 = f(x + 0.5 * k1) * dt
    k3 = f(x + 0.5 * k2) * dt
    k4 = f(x + k3) * dt
    return x + (k1 + 2 * k2 + 2 * k3 + k4) / 6.0

# Harmonic spring problem

y0 = 1
v0 = 0
k = 1
m = 1

target = 0.01
ddt = 0.0001 # Amount by which to adjust dt on each trial

def harmonic(x):
    return array([x[VELOCITY], -k * x[POSITION] / m])

def harmonic_energy(x):
    return 0.5 * k * x[POSITION] ** 2 + 0.5 * m * x[VELOCITY] ** 2

def true_harmonic(t):
    return y0 * cos(sqrt((k / m)) * t)

def simulate(integ, dt):
    state = [array([y0,v0])]
    energy = [harmonic_energy((y0, v0))]
    time = [0]
    while time[-1] < 30:
        state.append(integ(state[-1], harmonic, dt))
        energy.append(harmonic_energy(state[-1]))
        time.append(time[-1] + dt)
    return state, energy, time

# Find a dt that achieves the specified accuracy
for integrator in (euler, euler_richardson, runge_kutta):
    dt = 0.5
    while 1:
        state, energy, time = simulate(integrator, dt)

        e0 = energy[0]
        en = energy[-1]
    
        if abs(en - e0) / abs(e0) < target:
            # Compute the true data
        
            print 'Integrator: {}'.format(integrator.__name__)
            print 'Maximum time step to achieve {}% error: {}'.format(target * 100, dt)
            print 'Change in energy: {}'.format(energy[-1] - energy[0])

            #subplot(2,1,1)
            plot(time, [s[POSITION] for s in state], label='Simulated')
            plot(time, [true_harmonic(t) for t in time], label='Analytical')
            title('Harmonic Motion - Position')
            xlabel('Time (s)')
            ylabel('Position (m)')
            legend(loc='upper left')
            show()

            #subplot(2,1,2)
            plot(time, energy, label='Simulated')
            plot(time, [energy[0]] * len(energy), label='Analytical')
            ylim(min(energy) - 0.01, max(energy) + 0.01)
            title('Harmonic Motion - Energy')
            xlabel('Time (s)')
            ylabel('Energy (J)')
            legend(loc='lower left')
            show()

            break
        else:
            dt -= ddt
```
